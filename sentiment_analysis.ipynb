```python
data=pd.read_csv('/content/drive/MyDrive/Verzeo-Major Project/Restaurant_Reviews.tsv',sep='\t')
data.head()
```

```python
data.shape
```

```python
data['Liked']=data['Liked'].replace({1:'Positive',0:'Negative'})
data
```

```python
data.info()
```

```python
x=data['Review']
y=data['Liked']
```

```python
data['Liked'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.svm import SVC
from sklearn.pipeline import make_pipeline
sentiment_analyzer= make_pipeline(CountVectorizer(),SVC())
sentiment_analyzer.fit(x_train,y_train)
```

```python
y_pred=sentiment_analyzer.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)*100
```

```python
import joblib
joblib.dump(sentiment_analyzer,open('sentiment_analyzer_model.job','wb'))
```
